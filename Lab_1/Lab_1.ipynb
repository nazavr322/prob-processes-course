## 3. Обчислити для згенерованого масиву чисел вибіркове математичне сподівання $\overline{x}$ , вибіркову дисперсію $\sigma ^ 2$ , вибіркове середньоквадратичне відхилення $\sigma$. 

```python
# create figure and add title
axes = plt.subplots(2, 2, figsize=(10, 10), dpi=300)[1].flatten()
plt.suptitle('Experiment Results', fontsize=14)

# calculate theoretical values
a = np.arange(1, 9)
df = {'n': ['$\infty$'], 'm': [a.mean()], 'd': [a.var()], 's': [a.std()]}

for i in range(1, 5):
    n = 10 ** i
    # create array of random numbers
    a = np.random.randint(1, 9, n)
    df['n'].append(n)
    # calculate statistics
    df['m'].append(a.mean())
    df['d'].append(a.var(ddof=1))
    df['s'].append(a.std(ddof=1))
    # show plot
    ax = axes[i - 1]
    ax.set_title(f'Histogram of observations for $N = {n}$')
    sns.histplot(a, stat='probability', discrete=True, ax=ax)
sns.despine()
plt.tight_layout()

# create DataFrame with results
df = pd.DataFrame(df)
df.columns = ['$N$', '$\overline{x}$', '$\sigma^2$', '$\sigma$']
df
```

## 4. Зобразити графічно залежності $\overline{x}$ та $\sigma$ від $lg(N)$. Проаналізувати отримані результати.

```python
plt.title(f'Dependency of {df.columns[1]} and {df.columns[3]} for different values of $N$')
plt.plot(df.iloc[1:, 0], df.iloc[1:, 1], label=df.columns[1])
plt.plot(df.iloc[1:, 0], df.iloc[1:, 3], label=df.columns[3])
plt.xscale('log')
plt.legend()
plt.xlabel('$lg(N)$')
sns.despine()
```
