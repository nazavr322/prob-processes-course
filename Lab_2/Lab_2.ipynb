## 3. Розгляньте випадок 6-ти дайсів: 4d, 6d, 8d, 10d, 12d та 20d. Згенеруйте випадкові результати 10 кочень випадково вибраного дайсу і знайдіть ймовірності вибору кожного конкретного дайсу.

```python
bi = 1 / 6
dices = [4, 6, 8, 10, 12, 20]
rolls = np.random.randint(1, 21, 10)
bi_a = []
for roll in rolls:
    # calculate probability to appear on each dice (P(A | B_i))
    a_bi = np.array([1 / dice if roll <= dice else 0 for dice in dices])
    # calculate P(B_i)P(A|B_i)
    bi_a_bi = a_bi * bi
    # calculare P(B_i | A) and set it as new B_i
    bi = bi_a_bi / bi_a_bi.sum()
    bi_a.append(bi)
bi_a = np.array(bi_a)

# gather experiment results
df = {'Roll': rolls}
df.update({f'$P(B_{{{d}d}}|A)$': bi_a[:, i] for i, d in enumerate(dices)})
df = pd.DataFrame(df, index=range(1, len(rolls) + 1))

# plot results
plt.figure(figsize=(12, 6), dpi=300)
plt.title(f'Experiment results', fontsize=14)
plt.xlabel('Roll Number', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.xticks(range(1, 11), range(1, 11))
for i, col_name in enumerate(df.columns[1:], 1):
    plt.plot(df.iloc[:, i], label=col_name)
plt.legend();

df
```
