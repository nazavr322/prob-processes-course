## 1. Дослідити властивості одновимірної випадкової величини $Х_1$:
### 1.1 Згенерувати вибірку випадкової величини $Х_1$, яка містить не менше 30 значень при заданому середньоквадратичному відхиленні $\sigma_x = 15$
### 1.2 Побудувати варіаційний ряд
### 1.3 З допомогою варіаційного ряду побудувати діаграму накопичених частот
### 1.4 Побудувати гістограму вибірки
### 1.5 Розрахувати за формулами оцінки математичного сподівання, дисперсії та середньо квадратичного відхилення випадкової величини $Х_1$

```python
# generate 30 samples with std=15
x = np.random.normal(scale=15, size=30)
# create variational series
z = np.sort(x)

# plot the diagram of cumulative frequencies
plt.title('Diagram of cumulative frequencies')
plt.xlabel('$z$')
plt.ylabel('$\hat{F}_N(x)$')
plt.step(z, [(z < i).sum() / len(z) for i in z])
sns.despine()
plt.show()

# plot the histogram
plt.title('Histogram')
sns.histplot(x, bins=10, stat='density')
sns.despine()
plt.show()

# calculate E[X], Var[X] and s
print('Experimental E[X]:', x.mean())
print('Experimental Var[X]:', x.var(ddof=1))
print('Experimental s:', x.std(ddof=1))
```

## 2. Дослідити властивості двовимірної сукупності випадкових величин $Х_1$ та $Y_1$:
### 2.1 Згенерувати різні вибірки випадкових величин $Х_1$ та $Y_1$
### 2.3 Побудувати поля розсіяння
### 2.4 Скласти таблицю двовимірних розподілів
### 2.5 Порахувати коефіцієнти кореляції

```python
# generate N random samples in [-10, 10) interval
N = 100
x, y = np.random.randint(-10, 10, (2, N))

# plot scatterplot
plt.title(f'Scatterplot of {N} points')
plt.xlabel('$x$')
plt.ylabel('$y$')
sns.scatterplot(x=x, y=y)
sns.despine()
plt.show()

# create distribution table
H, xedges, _ = np.histogram2d(x, y)
idx = pd.IntervalIndex.from_breaks(xedges.round(1), 'left')
df = pd.DataFrame([list(zip(r, r / N)) for r in H], idx, idx)
print('2D Distribution Table:')
display(df)

# calculate correlation coefficient
p = (x - x.mean()).dot(y - y.mean()) / ((N - 1) * x.std(ddof=1) * y.std(ddof=1))
print('Correlation coefficient:', p)
```
